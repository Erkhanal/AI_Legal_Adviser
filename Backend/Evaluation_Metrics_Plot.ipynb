### 2. Radar / Spider Chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set Seaborn style
sns.set(style="whitegrid")

# Define performance metrics and models
metrics = ['Cosine Similarity', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'BLEU Score']
models = ['Gemini 2.5 Pro', 'GPT-3.5', 'BERT-Base']

# Data for English language
data_english = {
    'Metric': metrics,
    'Gemini 2.5 Pro': [86, 77, 95, 78, 86, 34],
    'GPT-3.5': [85, 70, 90, 73, 81, 14],
    'BERT-Base': [84, 70, 94, 68, 79, 16]
}

# Data for Nepali language
data_nepali = {
    'Metric': metrics,
    'Gemini 2.5 Pro': [89, 81, 96, 83, 89, 53],
    'GPT-3.5': [84, 70, 90, 73, 81, 20],
    'BERT-Base': [86, 66, 87, 71, 78, 26]
}

# Data for Romanized Nepali language
data_romanized_nepali = {
    'Metric': metrics,
    'Gemini 2.5 Pro': [87, 72, 88, 79, 83, 52],
    'GPT-3.5': [81, 58, 85, 63, 72, 18],
    'BERT-Base': [84, 66, 90, 67, 77, 33]
}

# Convert data to DataFrame 
df_english = pd.DataFrame(data_english)
df_nepali = pd.DataFrame(data_nepali)
df_romanized_nepali = pd.DataFrame(data_romanized_nepali)

# Function to plot radar chart 
def radar_chart(ax, df, language, color_palette):
    labels = df['Metric'].values
    values_gemini = df['Gemini 2.5 Pro'].values
    values_gpt = df['GPT-3.5'].values
    values_bert = df['BERT-Base'].values

    # Number of variables
    num_vars = len(labels)

    # Compute angle for each metric
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # Make the plot close to a circle
    values_gemini = np.concatenate((values_gemini, [values_gemini[0]]))
    values_gpt = np.concatenate((values_gpt, [values_gpt[0]]))
    values_bert = np.concatenate((values_bert, [values_bert[0]]))
    angles += angles[:1]

    # Plot on the given axes
    ax.fill(angles, values_gemini, color=color_palette[0], alpha=0.25)
    ax.fill(angles, values_gpt, color=color_palette[1], alpha=0.25)
    ax.fill(angles, values_bert, color=color_palette[2], alpha=0.25)

    ax.plot(angles, values_gemini, color=color_palette[0], linewidth=2, label='Gemini 2.5 Pro')
    ax.plot(angles, values_gpt, color=color_palette[1], linewidth=2, label='GPT-3.5')
    ax.plot(angles, values_bert, color=color_palette[2], linewidth=2, label='BERT-Base')

    # Add labels and title
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=10)
    ax.set_title(f'Base LLM Performance - {language}', size=14, color='black', fontweight='bold')
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), fontsize=10)

# Define color palettes 
english_palette = ['#8BAFDC', '#F1D4A1', '#A7D4B5']  
nepali_palette = ['#A1A9D8', '#C1D7A3', '#F1B4B0']  
romanized_nepali_palette = ['#8AC8D1', '#F3A573', '#D06A84']  

# Create a subplot for each language
fig, axes = plt.subplots(1, 3, figsize=(18, 6), subplot_kw=dict(polar=True))

# Plot for English language
radar_chart(axes[0], df_english, 'English', english_palette)

# Plot for Nepali language
radar_chart(axes[1], df_nepali, 'Nepali', nepali_palette)

# Plot for Romanized Nepali language
radar_chart(axes[2], df_romanized_nepali, 'Romanized Nepali', romanized_nepali_palette)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

### Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set Seaborn style
sns.set(style="whitegrid")

# Define performance metrics and models
metrics = ['Cosine Similarity', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'BLEU Score']
models = ['Gemini 2.5 Pro', 'GPT-3.5', 'BERT-Base']

# Data for English language
data_english = {
    'Metric': metrics,
    'Gemini 2.5 Pro': [86, 77, 95, 78, 86, 34], 
    'GPT-3.5': [85, 70, 90, 73, 81, 14],
    'BERT-Base': [84, 70, 94, 68, 79, 16]
}

# Data for Nepali language
data_nepali = {
    'Metric': metrics,
    'Gemini 2.5 Pro': [89, 81, 96, 83, 89, 53], 
    'GPT-3.5': [84, 70, 90, 73, 81, 20],
    'BERT-Base': [86, 66, 87, 71, 78, 26]
}

# Data for Romanized Nepali language
data_romanized_nepali = {
    'Metric': metrics,
    'Gemini 2.5 Pro': [87, 72, 88, 79, 83, 52], 
    'GPT-3.5': [81, 58, 85, 63, 72, 18],
    'BERT-Base': [84, 66, 90, 67, 77, 33]
}

# Convert data to DataFrame 
df_english = pd.DataFrame(data_english)
df_nepali = pd.DataFrame(data_nepali)
df_romanized_nepali = pd.DataFrame(data_romanized_nepali)

# Function to plot heatmap
def heatmap_chart(df, language, color_palette, ax):
    
    data = df.set_index('Metric').T  
    
    # Plot the heatmap
    sns.heatmap(data, annot=True, fmt='.1f', cmap=color_palette, cbar=True, annot_kws={"size": 14}, linewidths=0.5, vmin=0, vmax=100, ax=ax)

    # Title and labels
    ax.set_title(f'Base LLM Performance Heatmap - {language}', fontsize=16, fontweight='bold')
    ax.set_ylabel('Model', fontsize=12)
    ax.set_xlabel('Metric', fontsize=12)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=10)
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=10)


english_palette = 'YlGnBu'  
nepali_palette = 'coolwarm' 
romanized_nepali_palette = 'YlOrRd'  

# Create a 1x3 subplot grid
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot for English language 
heatmap_chart(df_english, 'English', english_palette, axes[0])

# Plot for Nepali language 
heatmap_chart(df_nepali, 'Nepali', nepali_palette, axes[1])

# Plot for Romanized Nepali language 
heatmap_chart(df_romanized_nepali, 'Romanized Nepali', romanized_nepali_palette, axes[2])

# Adjust layout 
plt.tight_layout()

# Show plots
plt.show()
```
